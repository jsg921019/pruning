```python
import matplotlib.pyplot as plt
import numpy as np
from numpy import ma
from matplotlib import ticker, cm

N = 100

time = np.linspace(40.0, 90.0, N)
f1 = np.linspace(0.5, 0.9, N)

time, f1 = np.meshgrid(time, f1)

t_score = 0.5 * (time / 43.16)
f_score = sigmoid(20 * (0.7 - f1))

mtx = t_score + f_score

fig, ax = plt.subplots(figsize=(15,10))

levs = np.linspace(0.3, 2.1, 19)
cs = ax.contourf(time, f1, mtx, levs, cmap=cm.PuBu)

cbar = fig.colorbar(cs)
size_coef = 5

ax.scatter([80.18], [0.7721], c='white', edgecolors='black', s=22*size_coef)
ax.text(80.18+1, 0.7721, 'Mobilenet v3', size=13, ha="left", va="center")

ax.scatter([52.65], [0.6885], c='white', edgecolors='black', s=234*size_coef)
ax.text(52.65+2, 0.6885, 'Alexnet', size=13, ha="left", va="center")

ax.scatter([62.48], [0.8114], c='white', edgecolors='black', s=504*size_coef)
ax.text(62.48+2.5, 0.8114, 'VGG11', size=13, ha="left", va="center")

ax.scatter([52.75], [0.7890], c='white', edgecolors='black', s=35.2*size_coef)
ax.text(52.75+1, 0.7890, 'VGG9', size=13, ha="left", va="center")

ax.scatter([47.64], [0.7817], c='white', edgecolors='black', s=35.2*size_coef)
ax.text(47.64+1, 0.7817-0.006, 'VGG9 res=128', size=13, ha="left", va="center")

ax.scatter([47.31], [0.7659], c='white', edgecolors='black', s=21*size_coef, label='vgg9 112 pruned')
ax.text(47.31+1, 0.7659-0.005, 'VGG9 pruned x0.75', size=13, ha="left", va="center")

ax.set_ylabel('F1 score')
ax.set_xlabel('Time')

plt.show()
```

```python
labels = ['MobilentV3', 'Shufflenet', 'Squeezenet', 'VGG', 'Alexnet']

cmap = plt.get_cmap("Set2")
# for i in range(4):
#     ax.plot(t,i*(t+1),   color=cmap(i), linestyle = '-')
#     ax.plot(t,i*(t+1)+.3,color=cmap(i), linestyle = ':')

tot = np.array([80, 77, 46.2, 39.7, 36.39])
load_data = np.array([13, 8, 14, 16, 19.75])
profile = np.array([22, 26, 12.6, 7.18, 5.54])
forward = np.array([44.5, 42.2, 19.5, 14.8, 10.13])

width = 0.35

fig, ax = plt.subplots(figsize=(10,6))

ax.bar(labels, tot, width, label='Etc', color=cmap(3))
ax.bar(labels, load_data, width, label='Data loading', color=cmap(0))
ax.bar(labels, profile, width, bottom=load_data, label='Profiling', color=cmap(1))
ax.bar(labels, forward, width, bottom=load_data+profile, label='Forward pass', color=cmap(2))


ax.set_ylabel('time')
ax.set_title('Time')
ax.legend()

plt.show()
```
